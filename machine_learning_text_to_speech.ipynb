```python
!pip install gTTS    # Download da biblioteca utilizada
from gtts import gTTS
from IPython.display import Audio
```

```python
text_to_say = "Do you like to play volleyball?"    # Variável para armazenar o texto

language = "en"

gtts_object = gTTS(text = text_to_say,
                  lang = language,
                  slow = False)
```

```python
gtts_object.save("/content/gtts.wav")    # salvando em um arquivo wav
```

```python
Audio("/content/gtts.wav")    # àudio com o texto que foi criado e a linguagem pedida
```

Testando com outro idioma (Portugues):

```python
ptbr_text = "Hoje o dia amanhaceu chuvoso"

ptbr_language = "pt"

ptbr_gtts_object = gTTS(text = ptbr_text,
                        lang = ptbr_language,
                        slow = True)
```
